```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


alpha=0.33
beta=0.98

L0=np.asmatrix(np.zeros((2,2)))
L1=np.asmatrix(np.zeros((2,2)))

L0[0,0]=-alpha+1
L0[0,1]=1
L0[1,0]=1

L1[0,1]=1
L1[1,0]=1/beta
L1[1,1]=-(1/(alpha*beta)-1)

if np.linalg.det(L0)!=0:
    print('the matrix L0 is invertible')
    A=np.linalg.inv(L0)*L1
    print(A)
else:
    print('the matrix L0 is not invertible')
```

```python
####################################
# CALCULATE EIGNEVALUES
####################################
#DEFINE AS AS IN THE NOTES
A=np.linalg.inv(L0)*L1

#%%
w,ve=np.linalg.eig(A)

#w are the eigenvalues
#ve is the matrix of eigenvector
#%%
#Create matrix of eigenvalues
MU=np.asmatrix(np.zeros((2,2)))
MU[0,0]=w[0]
MU[1,1]=w[1]

#SHOULD BE AS A
AP=ve*MU*np.linalg.inv(ve)

#Check Diagonlization
DD=AP-A
DD=abs(DD)
if DD.max()<10**(-5):
    print("Matrix A has been diagonalized")
else:
    print("Diagonalization failed")
    
#trace of A
traceA=A[0,0]+A[1,1]

#Determinant of A
detA=A[0,0]*A[1,1]-A[0,1]*A[1,0]

#%%
#Check linear Algebra Properties
if detA-w[0]*w[1]<10**(-5):
    print ("determinant A is product of eigenvalue")
else:
    print ("Linear Algebra properties not satisfied")        
     
if traceA-(w[0]+ w[1])<10**(-5):
    print ("Trace of  A is sum of eigenvalue")
else:
    print ("Linear Algebra properties not satisfied")        
```

```python
eigenValues,eigenVectors = np.linalg.eig(A)
#eigenValues is w 
#eigenVectors is v

idx = eigenValues.argsort()[::-1]  # return the index of  largest to smallest 
eigenValues = eigenValues[idx]  #sort from largest to smallets
eigenVectors = eigenVectors[:,idx]  #sort eigenvectors

if eigenValues[1] >= 1: #smallest larger than one
    print ("no saddle path, both eigenvalues larger than one")
elif eigenValues[0] <= 1: #largest smaller than
    print ("no saddle path, both eigenvalues smaller than one")
else:
    print  ("Saddle Path passed")
    print ("One eigenvalue is greater that one and one is lower than one") 
    print (eigenValues)
```

```python
# Import numpy, pandas, linearsolve, matplotlib.pyplot
import numpy as np
import pandas as pd
import linearsolve as ls
import matplotlib.pyplot as plt
plt.style.use('classic')

# Input model parameters
parameters = pd.Series(dtype=float)
parameters['alpha']  = 0.33
parameters['beta']  = 0.98

# Funtion that evaluates the equilibrium conditions
def equations(variables_forward,variables_current,parameters):

    # Parameters
    p = parameters

    # Variables
    fwd = variables_forward
    cur = variables_current

    # Household Euler equation
    euler_eqn = p.alpha*fwd.k**(p.alpha - 1)- fwd.c / (p.beta * cur.c)

    # Goods market clearing
    constraints= fwd.k + cur.c - (cur.k**p.alpha)

    # Stack equilibrium conditions into a numpy array
    return np.array([
            euler_eqn,
            constraints
        ])

# Initialize the model
model = ls.model(equations = equations,
                 n_states=1,
                 n_exo_states = 1,
                 var_names=['k','c'],
                 parameters = parameters)

guess = [0.2,0.4]
model.compute_ss(guess)

model.approximate_and_solve()
     

print("Transition function:",model.f)
print("Policy function:", model.p)
print(model.eig)

```
